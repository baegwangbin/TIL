# Correlation
># 1. Multivariate Discrete Random Variable

### Joint Prabability Mass Function

$$ P_{XY}(x, y) $$

### Marginal Probability Mass Function
* Integration of joint pmf

$$ 
\begin{align}%\label{}
\nonumber  P_X(x) = \sum_{y_i} f_{XY}(x,y_i) \\
\nonumber  P_Y(y) = \sum_{x_i} f_{XY}(x_i,y) \\
\end{align} 
$$

### Conditional Probability Mass Function
* Cross-section of joint pmf

$$ P_{X \mid Y}(x \mid y) = \dfrac{P_{XY}(x, y)}{P_{Y}(y)} $$

$$ P_{Y \mid X}(y \mid x) = \dfrac{P_{XY}(x, y)}{P_{X}(x)} $$

># 2. Multivariate Continuous Random Variable

### Joint Cumulative Porbability Distribution

$$ F_{XY}(x, y) = P(\{ X < x \} \cap \{ Y < y \}) = P(X < x, Y < y) $$

### Marginal Cumulative Probability Distribution

$$ F_X(x)=F_{XY}(x, \infty) $$

$$ F_Y(y)=F_{XY}(\infty, y) $$

$$ F_{XY}(\infty, \infty)=1 $$

$$ F_{XY}(-\infty, y)=F_{XY}(x,-\infty)=0 $$

### Joint Probability Density Function

$$ f_{XY} = \dfrac{\partial^2 F_{XY}(x, y)}{\partial x \partial y} $$

$$  \int_{x_1}^{x_2} \int_{y_1}^{y_2} f_{XY}(x,y)dxdy = P\big(\{ x_1 \leq X \leq x_2, \; y_1 \leq Y \leq y_2 \}\big)  $$

$$  \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} f_{XY}(x,y)dxdy=1 $$

### Marginal Probability Density Function

$$ 
\begin{align}%\label{}
\nonumber  f_X(x) = \int_{-\infty}^{\infty} f_{XY}(x,y)dy \\
\nonumber  f_Y(y) = \int_{-\infty}^{\infty} f_{XY}(x,y)dx
\end{align} 
$$

### Conditional Probability Density Function

$$ f_{X \mid Y}(x \mid y_0) = \dfrac{f_{XY}(x, y=y_0)}{f_{Y}(y_0)} $$

$$ f_{Y \mid X}(y \mid x_0) = \dfrac{f_{XY}(x=x_0, y)}{f_{X}(x_0)} $$

># 3. Independence

### Condition for Independence

$$ f_{XY}(x, y) = f_X(x)f_Y(y) $$

### Repetition

* Random Variable: $f(x)$
* Sample Data: $\{x_1, x_2, x_3, \cdots, x_N \}$
* Probability

$$ f(x_1, x_2, x_3, \cdots, x_N ) = \prod_{i=1}^N f(x_i) $$  
 

### Expectation Values

$$ \text{E}[XY] = \text{E}[X]\text{E}[Y] $$

$$ \text{E}[(X - \mu_X)(Y-\mu_Y)] = 0 $$

Proof)

$$ 
\begin{eqnarray}
\text{E}[XY] 
&=& \int xy \;f_{XY}(x, y) \; dx dy  \\
&=& \int xy \;f_{X}(x)f_{Y}(y) \; dx dy \\
&=& \int x \;f_{X}(x) \; dx \int y \;f_{Y}(y) \; dy  \\
&=& \text{E}[X] \text{E}[Y] \\
\end{eqnarray}
$$

$$ 
\begin{eqnarray}
\text{E}[(X - \mu_X)(Y-\mu_Y)] 
&=& \text{E}\left[XY - \mu_XY-\mu_YX + \mu_X\mu_Y\right] \\
&=& \text{E}[XY] - \mu_X\text{E}[Y]-\mu_Y\text{E}[X] + \mu_X\mu_Y  \\
&=& \text{E}[XY] - \mu_X\mu_Y  \\
&=& \text{E}[XY] - \text{E}[X]\text{E}[Y] = 0 
\end{eqnarray}
$$

### Variance

$$ \text{Var}[X+Y] = \text{Var}[X] + \text{Var}[Y] $$

proof)

$$ 
\begin{eqnarray}
\text{Var}[X + Y] 
&=& \text{E}[((X + Y) - (\mu_X + \mu_Y))^2] \\
&=& \text{E}[(X+Y)^2 - 2(X+Y)(\mu_X + \mu_Y) + (\mu_X + \mu_Y)^2] \\
&=& \text{E}[X^2+2XY+Y^2] - 2(\mu_X + \mu_Y)\text{E}[X+Y] + (\mu_X + \mu_Y)^2 \\
&=& \text{E}[X^2+2XY+Y^2] - 2(\mu_X + \mu_Y)^2 + (\mu_X + \mu_Y)^2 \\
&=& \text{E}[X^2]+2\text{E}[XY]+\text{E}[Y^2] - (\mu_X + \mu_Y)^2 \\
&=& \text{E}[X^2]+2\text{E}[X]\text{E}[Y]+\text{E}[Y^2] - (\mu_X^2 + 2\mu_X\mu_Y + \mu_Y^2) \\
&=& \text{E}[X^2]-\mu_X^2+\text{E}[Y^2]-\mu_Y^2+2\text{E}[X]\text{E}[Y] - 2\mu_X\mu_Y \\
&=& \text{Var}[X]+\text{Var}[Y] \\
\end{eqnarray}
$$

### Conditional Probability Distribution

$$ f_{X \mid Y} (x | y_0) = \dfrac{f_{XY}(x, y=y_0)}{f_{Y}(y_0)} = \dfrac{f_{X}(x) f_{Y}(y_0)}{f_{Y}(y_0)} = f_{X}(x) $$

$$ f_{Y \mid X} (y | x_0) = \dfrac{f_{XY}(x=x_0, y)}{f_{X}(x_0)} = \dfrac{f_{X}(x_0) f_{Y}(y)}{f_{X}(x_0)} = f_{Y}(y) $$

># 4. Covariance & Correlation Coefficient

### Sample Covariance

* $m_x$: sample average of $x$
* $m_y$: sample average of $y$

$$ s_{xy} = \dfrac{1}{N}\sum_{i=1}^{N} (x_i-m_x)(y_i-m_y) $$

### Sample Correlation Coefficient

* Normalized Covariance

$$ r_{xy} = \dfrac{s_{xy}}{\sqrt{s^2_{x} \cdot s^2_{y}}} $$

### Covariance of Random Variables

$$ \text{Cov}[X, Y] = \text{E}[(X - \text{E}[X])(Y - \text{E}[Y])]$$

### Correlation Coefficient of Random Variables

$$ \rho[X,Y] =  \dfrac{\text{Cov}[X, Y]}{\sqrt{\text{Var}[X] \cdot \text{Var}[Y]}} $$

$$ -1 \leq \rho \leq 1 $$

* $\rho = 1$ : total positive linear correlation
* $\rho = 0$ : no linear correlation (different from independent)
* $\rho = −1$ : total negative linear correlation

### Covariance of Multivariate Random Variables

* Sample Covariance Matrix

$$
S =
\begin{bmatrix}
\begin{eqnarray}
s_{x_1}^2     \;\;  &  s_{x_1x_2} \;\;&  s_{x_1x_3} \;\;&  \cdots &  s_{x_1x_N} \\
s_{x_1x_2}   \;\;    &  s_{x_2}^2 \;\;&  s_{x_2x_3} \;\;&  \cdots &  s_{x_2x_N} \\
\vdots       &  \vdots &  \vdots &  \ddots &  \vdots \\
s_{x_1x_N}   \;\;    &  s_{x_2x_N} \;\;&  s_{x_3x_N} \;\;&  \cdots &  s_{x_N}^2 \\
\end{eqnarray}
\end{bmatrix}
$$

* $X_0$: zero-mean feature matrix (average subtracted)
* $N$: No. of Data

$$ S = \dfrac{1}{N} X_0^TX_0 $$

* Theoretical Covariance Matrix

$$ \Sigma = \text{Cov}[X] = \text{E} \left[ (X - \text{E}[X])(X - \text{E}[X])^T \right] =
\begin{bmatrix}
\begin{eqnarray}
\sigma_{x_1}^2     \;\;  &  \sigma_{x_1x_2} \;\;&  \sigma_{x_1x_3} \;\;&  \cdots &  \sigma_{x_1x_N} \\
\sigma_{x_1x_2}   \;\;    &  \sigma_{x_2}^2 \;\;&  \sigma_{x_2x_3} \;\;&  \cdots &  \sigma_{x_2x_N} \\
\vdots       &  \vdots &  \vdots &  \ddots &  \vdots \\
\sigma_{x_1x_N}   \;\;    &  \sigma_{x_2x_N} \;\;&  \sigma_{x_3x_N} \;\;&  \cdots &  \sigma_{x_N}^2 \\
\end{eqnarray}
\end{bmatrix}$$

># 5. Multivariate Normal
* MVN(Multivariate Normal): Most widely used distribution to model multiple random variables

$$ \mathcal{N}(x ; \mu, \Sigma) = \dfrac{1}{(2\pi)^{D/2} |\Sigma| ^{1/2}} \exp \left( -\dfrac{1}{2} (x-\mu)^T \Sigma^{-1} (x-\mu) \right) $$

* $D$: No. of Dimension
* $x \in \mathbf{R}^D $ Random Variable Vector
* $\mu \in \mathbf{R}^D $ Average Vector
* $\Sigma \in \mathbf{R}^{D\times D} $ Covariance Matrix
* $\Sigma^{-1} \in \mathbf{R}^{D\times D} $ Precision Matrix or Concentration Matrix (Inverse of Covariance Matrix)

### MVN in SciPy
* `sp.stats.multivariate_normal(mean, cov)`
* `mean`: average vector
* `cov`: covariance matrix

### example

$$ 
\mu = \begin{bmatrix}2 \\ 3 \end{bmatrix}. \;\;\;
\Sigma = \begin{bmatrix}1 & 0 \\ 0 & 1 \end{bmatrix} 
$$

$$ 
| \Sigma| = 1. \;\;\;
\Sigma^{-1} = \begin{bmatrix}1 & 0 \\ 0 & 1 \end{bmatrix}
$$

$$ 
(x-\mu)^T \Sigma^{-1} (x-\mu) =  
\begin{bmatrix}x_1 - 2 & x_2 - 3 \end{bmatrix}
\begin{bmatrix}1 & 0 \\ 0 & 1 \end{bmatrix} 
\begin{bmatrix}x_1 - 2 \\ x_2 - 3 \end{bmatrix}
=
(x_1 - 2)^2 + (x_2 - 3)^2
$$

$$
\mathcal{N}(x_1, x_2) = \dfrac{1}{2\pi}
\exp \left( -\dfrac{1}{2} \left( (x_1 - 2)^2 + (x_2 - 3)^2 \right) \right)
$$

```python
mu = [2, 3]
cov = [[1, 0], [0, 1]]
rv = sp.stats.multivariate_normal(mu, cov)
xx = np.linspace(0, 4, 120)
yy = np.linspace(1, 5, 150)
XX, YY = np.meshgrid(xx, yy)
plt.grid(False)
plt.contourf(XX, YY, rv.pdf(np.dstack([XX, YY])))
plt.axis("equal")
plt.show()
```
