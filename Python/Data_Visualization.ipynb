### 1.2. Customize Style
* Style string: **color + marker + line style**
* **color**
  * RGB or RGBA tuple of float values in [0, 1]
  * hex RGB or RGBA tuple of float values in [0, 1]
  * string representation of a float value in [0, 1] $\rightarrow$ gray level
  * one of {**'b'**lue, **'g'**reen, **'r'**ed, **'c'**yan, **'m'**agenta, **'y'**ellow, blac**'k'**, **'w'**hite};

* **marker**

| string | marker |
|-|-|
| `.` | point marker  |
| `,` | pixel marker |
| `o` | circle marker |
| `v` | triangle_down marker |
| `^` | triangle_up marker |
| `<` | triangle_left marker |
| `>` | triangle_right marker |
| `1` | tri_down marker |
| `2` | tri_up marker |
| `3` | tri_left marker |
| `4` | tri_right marker |
| `s` | square marker |
| `p` | pentagon marker |
| `*` | star marker |
| `h` | hexagon1 marker |
| `H` | hexagon2 marker |
| `+` | plus marker |
| `x` | x marker |
| `D` | diamond marker |
| `d` | thin_diamond marker |

* **line style**

| string | line style |
|-|-|
| `-` |  solid line style
| `--` |  dashed line style
| `-.` |  dash-dot line style
| `:` |  dotted line style

* **to add more style, specify style parameters**

| parameter | style |
|-|-|
| `c`  | `color` | 
| `lw` | `linewidth` |
| `ls` | `linestyle` |
| `marker` | `marker` |
| `ms` | `markersize` |
| `mec` | `markeredgecolor` |
| `mew` | `markeredgewidth` |
| `mfc` | `markerfacecolor` |



````python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')
plt.show()
````

````python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b", 
         lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.show()
````

### 1.3. Customize Range / Ticks / Tick Lables / Grid

````python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 
         c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")

# Range
plt.xlim(0, 50)
plt.ylim(-10, 30)

# Ticks
plt.xticks([10, 20, 30, 40])

# Tick Labels
plt.yticks([0, 10, 20], ["Low", "Zero", "High"])

# Grid
plt.grid(True)


plt.show()
````

### 1.4. Legend
* `loc` parameter: specifies the location of the legend

| loc | location |
| - |-|
| 0 | `best` |
| 1 | `upper right` |
| 2 | `upper left` | 
| 3 | `lower left` | 
| 4 | `lower right` | 
| 5 | `right` |
| 6 | `center left` |
| 7 | `center right` |
| 8 | `lower center` |
| 9 | `upper center` |
| 10 | `center` | 

````python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, ls="--", label="cosine")
plt.plot(X, S, ls=":", label="sine")
plt.legend(loc=2)
plt.show()
````

### 1.5. Axis Labels & Plot Title

````python
X = np.linspace(-np.pi, np.pi, 256)
C, S = np.cos(X), np.sin(X)
plt.plot(X, C, label="cosine")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.title("Cosine Plot")
plt.show()
````

### 1.6. Multiple Plots
* Single `plot` command
  * `plt.plot(x1, y1, style1, x2, y2, style2, ...)`
  * In this case, x data & style string cannot be omitted
* Multiple `plot` command
  * use `plt.hold(True)` for Matplotlib 1.5 or less

````python
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.show()

# or equivalently
plt.plot(t, t, 'r--')
plt.plot(t, 0.5 * t**2, 'bs:')
plt.plot(t, 0.2 * t**3, 'g^-')
plt.show()
````

### 2.2. Axes Object & `subplot` Command
* Use `tight_layout` to automatically adjust the gap between the plots
* Example
```
subplot(2, 1, 1)
# commands for upper plot
subplot(2, 1, 2)
# commands for lower plot
```
* OR equivalently
```
subplot(211)
# commands for upper plot
subplot(212)
# commands for lower plot
```



#### Example 1 - Two Plots

````python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

ax1 = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
````

#### Example 2 - Four Plots

````python
np.random.seed(0)

plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title("axes 1")

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title("axes 2")

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title("axes 3")

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title("axes 4")

plt.tight_layout()
plt.show()
````

#### Example 3 - generate multiple `Axes` objects using `subplots` command

````python
fig, axes = plt.subplots(2, 2)

np.random.seed(0)
axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title("axes 1")
axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title("axes 2")
axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title("axes 3")
axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title("axes 4")

plt.tight_layout()
plt.show()
````

### 2.3. Axis Object
* Use `twinx` command to generate a new `Axes` object with shared `x axis`

````python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()

ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)

plt.show()
````

### 3.8. Contour Plot
* 3D data $\rightarrow$ x, y, height
* `contour`: contour only
* `contourf`: contour + color
  * `meshgrid` generates grid point matrix

````python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)
````

````python
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
````

### 3.9. 3D Surface Plot

````python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)

XX, YY = np.meshgrid(X, Y)

RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)
````

````python
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
````

### 4.2. Generate

````python
x = np.asarray([0, 1, 2])
y = np.asarray([0, np.sqrt(3), 0])
triangles = [[0, 1, 2]]
triang = mtri.Triangulation(x, y, triangles)
plt.triplot(triang, 'ko-')
plt.xlim(-0.1, 2.1)
plt.ylim(-0.1, 1.8)
plt.show()
````

````python
x = np.asarray([0, 1, 2, 3, 4, 2])
y = np.asarray([0, np.sqrt(3), 0, np.sqrt(3), 0, 2*np.sqrt(3)])
triangles = [[0, 1, 2], [2, 3, 4], [1, 2, 3], [1, 3, 5]]
triang = mtri.Triangulation(x, y, triangles)
plt.triplot(triang, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.show()
````

### 4.3. Refine
* `TriRefiner` & `UniformTriRefiner` class
  * `refine_triangulation`: simple refine
  * `refine_field`: optimized refine

````python
refiner = mtri.UniformTriRefiner(triang)
triang2 = refiner.refine_triangulation(subdiv=2)
plt.triplot(triang2, 'ko-')
plt.xlim(-0.1, 4.1)
plt.ylim(-0.1, 3.7)
plt.show()
````

### 4.4. Contour Plot
* `tricontour` or `tricontourf` command

````python
triang5 = refiner.refine_triangulation(subdiv=5)
z5 = np.cos(1.5*triang5.x)*np.cos(1.5*triang5.y)
plt.tricontourf(triang5, z5)
plt.show()
````

### 4.5. Surface Plot
* `plot_trisur` or `plot_trisurf` command

````python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

triang3 = refiner.refine_triangulation(subdiv=3)
z3 = np.cos(1.5 * triang3.x) * np.cos(1.5 * triang3.y)

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(triang3.x, triang3.y, z3, cmap=cm.jet, linewidth=0.2)
ax.tricontourf(triang3, z3, zdir='z', offset=-1.2, cmap=cm.coolwarm)
ax.set_zlim(-1, 1)
ax.view_init(40, -40)
plt.show()
````
