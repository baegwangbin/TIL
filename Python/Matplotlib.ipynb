# Matplotlib
* easy-to-use data visualization tool

```python
import matplotlib.pyplot as plt
#display on jupyter notebook
%matplotlib inline 
```

```python
x = np.linspace(0,10,10)
y = x ** 2 + 2 * x + 1
```

># 1. simple plots

### .plot

plt.plot(x,y* default: line / ro: red circle / b^: blue triangle / ...
* '#e27f1b': RGB color index in hexadecimal* default: line / ro: red circle / b^: blue triangle / ...
* '#e27f1b': RGB color index in hexadecimal)

### .scatter
* **c**: color
* **marker**: type of marker = 'o', '^', '-'
* **alpha**: transparency

```python
plt.scatter(x,y, c='r', marker='o', alpha=0.5)
```

># 2. "figure" object
* **"figure"**: think of it as an empty canvas
* [more about figure](https://matplotlib.org/api/figure_api.html)

```python
fig = plt.figure()
axes2 = fig.add_axes([0.1, 0.1, 1.0, 1.0], projection='polar') # left, bottom, width, height
axes2.set_title("Here's a flower!")
x2 = list(range(0,100,1))
y2 = list(range(0,100,1))
axes2.plot(x2, y2, color='#ff0000', linestyle='-', linewidth=4)
```

### add_axes

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1.0, 1.0]) # left, bottom, width, height

x1 = np.array(range(0,100,1))
y1 = x1 ** 4
axes.plot(x1, y1)
# 
axes.set_xlabel('X values')
axes.set_ylabel('Y values')
axes.set_title('X-Y relationship')
```

### figsize & dpi
* **figsize**: size in inches
* **dpi**: dots per inch

```python
fig = plt.figure(figsize=(6, 3), dpi=100)
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([1.2, 0, 1, 1])

axes1.plot(x, y)
axes2.scatter(y, x)
```

># 3. customize plots

### axis label & scope

```python
plt.xlabel('X values')
plt.ylabel('Y values')

plt.xlim(0, 10) 
plt.ylim(0, 120)

plt.title('X-Y relationship')
plt.grid(True)

plt.plot(x, y)
```

### legend
* **label**
* **loc**: 'best' / 'center' / 'upper left' / ... 

```python
fig, axes = plt.subplots(figsize=(8, 6))

axes.set_xlim(0, 6)
axes.set_ylim(0, 60)

axes.plot(x, y, label='square')
axes.plot(x, x, label='identical')

axes.legend(loc='upper left')
```

### plot design (keyword parameters)
* **color** = `'r'`: red / `'b'`: blue / `'k'`: black/ `'#e27f1b'`: hexadecimal RGB index
* **linestyle** = `'-' / '--' / '-.' / ':' / ':.'`
* **linewidth**
* **marker** = `'o'`: circle / `'^'`: triangle
* **markerfacecolor**
* **markersize**
* [more about plot design](https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot)

```python
plt.plot(x, y, 'o', markerfacecolor='#ff0000', markersize = 30)
plt.plot(x, y, color='#0000ff', linewidth = 10)
```

># 4. subplot

### using simple plots

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(2, 2, 2)
plt.plot(y, x, 'g')

plt.subplot(2, 2, 3)
plt.plot(x, 2*x, 'b')

plt.subplot(2, 2, 4)
plt.plot(x, np.exp(x), 'k')
```

### using figure object - subplots()

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout() # prevent overlap
```

># 5. histogram

### hist()
* **data**
* **bins**: no. of separation
* **normed**: if True, normalize result

```python
data = np.random.randint(0, 101, size=500)
plt.hist(data, bins=10, alpha=0.5, normed=True)
plt.grid(True)
```
